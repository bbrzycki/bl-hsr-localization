What's a good distribution for choosing signal intensities? Let's use a very general proxy inspired by the empirical distribution found by Price et al. 2019. Here's a simple curve that approximates the fall-off at higher SNRs:

```python
x = np.linspace(10, 50, num=100, endpoint=False)
y = 1 - (x/50)**2
y1 = np.power(10, y)

plt.plot(x, np.log10(y1))
plt.xlabel('SNR (dB)')
plt.ylabel(r'$\log_{10}{\mathrm{hits}}$ (arbitrary units)')
plt.show()
```

Creating a probability distribution from this curve:

```python
xk = x
pk = y1 / np.sum(y1)

R = np.random.choice(xk, p=pk, size=10000)
plt.hist(R, bins=10)
plt.xlabel('SNR (dB)')
plt.ylabel(r'hits')
plt.yscale('log')
plt.show()
```

Generate a random, constant intensity signal with setigen:

```python
frame = stg.Frame(fchans=512,
                  tchans=16,
                  df=2.7939677238464355*u.Hz,
                  dt=18.25361108*u.s,
                  fch1=6095.214842353016*u.MHz)

frame.add_noise_from_obs(share_index=True)
noise_mean, noise_std = frame.get_noise_stats()

start_index = np.random.randint(0, fchans)
end_index = np.random.randint(0, fchans)
drift_rate = frame.get_drift_rate(start_index, end_index)

snr = np.power(10, np.random.choice(xk, p=pk) / 10)
width = np.random.uniform(10, 30)

signal = frame.add_constant_signal(f_start=frame.get_frequency(start_index),
                          drift_rate=drift_rate*u.Hz/u.s,
                          level=frame.get_intensity(snr=snr),
                          width=width*u.Hz,
                          f_profile_type='gaussian')
frame.render()

frame_info = {
    'noise_mean': noise_mean,
    'noise_std': noise_std,
    'snr': snr,
    'start_index': start_index,
    'end_index': end_index,
    'width': width
}
frame_info
```
